### Import the packages

```python
import cv2
from matplotlib import pyplot as plt

%matplotlib inline
```

### Let's download the images of two classical artists on which we will perform the face detection.

```python
!mkdir -p images
!wget \
    -O images/Edvard-Grieg.jpg \
    "https://upload.wikimedia.org/wikipedia/commons/c/c0/Edvard_Grieg_%281888%29_by_Elliot_and_Fry_-_02.jpg"

!wget \
    -O images/Giuseppe-Verdi.jpg \
    "https://m.media-amazon.com/images/M/MV5BYzM0ZDQxNjUtZDI1Ny00YTEyLWEzNmMtMTY3MjBmNTRlZjdjXkEyXkFqcGdeQXVyNTUyMTMyMDg@._V1_.jpg"
```

### Let's have a look at the two images

```python
eg_img = cv2.imread('images/Edvard-Grieg.jpg', 0)  # 0 is to convert the image to grayscale
print(eg_img.shape)  # Make sure it shows the image shape (rows, columns)

plt.imshow(eg_img, cmap='gray')
```

```python
gv_img = cv2.imread('images/Giuseppe-Verdi.jpg', 0)  # 0 is to convert the image to grayscale
print(gv_img.shape)  # Make sure it shows the image shape (rows, columns)

plt.imshow(gv_img, cmap='gray')
```

### Let's download the trained cascade classifier found at https://github.com/opencv/opencv/tree/master/data/haarcascades

### They have variety of classifiers. We download the frontalface classifier to detect faces in images.

```python
!mkdir -p features
!wget \
    -O features/haarcascade_frontalface_default.xml \
    "https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml"
```

### Load the face cascade classifier

```python
face_cascade = cv2.CascadeClassifier('features/haarcascade_frontalface_default.xml')
```

### Detect Faces

```python
faces = face_cascade.detectMultiScale(eg_img, 1.3, 5)
print(faces)  # Face Detection returns x and y coordinates and the height and width of rectangle
```

### Plot a bounding box around face

```python
for (x, y, w, h) in faces:
    eg_img_w_face = cv2.rectangle(eg_img, (x, y), (x + w, y + h), (255, 0, 0), 3)
    
plt.imshow(eg_img_w_face, cmap='gray')
```

### Very simple, isn't it! Within 5 lines of code :)

### Do the same for our second image

```python
faces = face_cascade.detectMultiScale(gv_img, 1.3, 5)
print(faces)  # Face Detection returns x and y coordinates and the height and width of rectangle

for (x, y, w, h) in faces:
    gv_img_w_face = cv2.rectangle(gv_img, (x, y), (x + w, y + h), (255, 0, 0), 3)
    
plt.imshow(gv_img_w_face, cmap='gray')
```

### You can utilise other haar cascade classifiers provided by OpenCV to detect eyes, full body, or even license plate numbers!
